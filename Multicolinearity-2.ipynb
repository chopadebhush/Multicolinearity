```python
df1 =df.copy()
```

```python
df1.drop(["total"],axis=1,inplace=True)
vif =pd.DataFrame()
vif["Feature"] =df1.columns
vif["VIF"] =[variance_inflation_factor(df1.values ,i) for i in range(len(df1.columns))]
vif
```

```python
def Feature_Selection(Dataframe ,drop_Col):
    Dataframe.drop(drop_Col,axis=1 ,inplace =True)
    vif =pd.DataFrame()
    vif["Feature"] =Dataframe.columns
    vif["VIF"] =[variance_inflation_factor(Dataframe.values ,i) for i in range(len(Dataframe.columns))]
    return vif

Feature_Selection(df1,"no_previous")
```

```python
Feature_Selection(df1 ,"ins_losses")
```
